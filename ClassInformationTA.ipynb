Getting Pipeline & Context

```python
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)

context_file = open('classInfo.txt','r')
```

```python
QA_Input = {'question': input('Your question (If you want to stop, type \'stop\'): '),
            'context': context_file.read()}

while(QA_Input['question'] != 'stop'):
  print('----------------------------------------------------------------------')
  print('Question: ' + QA_Input['question'])
  if (nlp(QA_Input)['score'] > 0.3): print('Answer: ' + nlp(QA_Input)['answer'])
  else: print('Answer: I am not too confident on that question, please contact the TA for this class if you need more information')
  QA_Input['question'] = input('Your question (If you want to stop, type \'stop\'): ')
```
